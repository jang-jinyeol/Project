```python
from retinaface import RetinaFace
import cv2
import matplotlib.pyplot as plt
```

```python
img_path="test.jpg"
```

```python
img = cv2.imread(img_path)
```

```python
obj = RetinaFace.detect_faces(img_path)
```

```python
obj
```

```python
obj.keys()
```

```python
len(obj.keys())
```

```python
face_area = obj['face_1']['facial_area']
```

```python
face_area
```

```python
cv2.rectangle(img,(face_area[2],face_area[3]),(face_area[0],face_area[1]),(0,255,0),3)
```

```python
plt.figure(figsize=(20,20))
plt.imshow(img[:,:,::-1])
plt.show()
```
